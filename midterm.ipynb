```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import pickle

df = pd.read_csv("heart.csv")
df.columns
df.columns = df.columns.str.lower().str.replace(' ','_')
df = df.fillna(0)
categorical = df.columns[df.dtypes == object]
numerical = df.columns[df.dtypes != object]


df_full_train ,df_test = train_test_split(df,test_size=0.2,random_state=1)
df_train ,df_val = train_test_split(df_full_train,test_size=0.25,random_state=1)
df_full_train = df_full_train.reset_index(drop=True)
df_test_train = df_test.reset_index(drop=True)
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
y_train = df_train.heartdisease.values
y_val = df_val.heartdisease.values
y_test = df_test.heartdisease
del df_train['heartdisease']
del df_val['heartdisease']
del df_test['heartdisease']


dv = DictVectorizer(sparse=False)
train_dict = df_train.to_dict(orient='records')
val_dict = df_val.to_dict(orient='records')
test_dict = df_test.to_dict(orient='records')
x_train = dv.fit_transform(train_dict)


model = LogisticRegression(solver='liblinear',C=1.0,random_state=1)
model.fit(x_train,y_train)

tree = DecisionTreeClassifier(max_depth=10,random_state=1)
tree.fit(x_train,y_train)

random = RandomForestClassifier(n_estimators=6,random_state=1)
random.fit(x_train,y_train)

scores_logistic = {}
depth_score = []
tree_score = {}
estimator_socre = []
forest_score ={}




y_pred_train = model.predict_proba(x_train)[:,1]
auc = roc_auc_score(y_train,y_pred_train)
scores_logistic['train'] = auc
x_val = dv.transform(val_dict)
y_pred_val = model.predict_proba(x_val)[:,1]
auc = roc_auc_score(y_val,y_pred_val)
scores_logistic['val'] = auc
x_test = dv.transform(test_dict)
y_pred_test = model.predict_proba(x_test)[:,1]
auc = roc_auc_score(y_test,y_pred_test)
scores_logistic['test'] = auc


y_pred_train = tree.predict_proba(x_train)[:,1]
auc = roc_auc_score(y_train,y_pred_train)
tree_score['train'] = auc
y_pred_val = tree.predict_proba(x_val)[:,1]
auc = roc_auc_score(y_val,y_pred_val)
tree_score['val'] = auc
y_pred_test = tree.predict_proba(x_test)[:,1]
auc = roc_auc_score(y_test,y_pred_test)
tree_score['test'] = auc


y_pred_train = random.predict_proba(x_train)[:,1]
auc = roc_auc_score(y_train,y_pred_train)
forest_score['train'] = auc
y_pred_val = random.predict_proba(x_val)[:,1]
auc = roc_auc_score(y_val,y_pred_val)
forest_score['val'] = auc
y_pred_test = random.predict_proba(x_test)[:,1]
auc = roc_auc_score(y_test,y_pred_test)
forest_score['test'] = auc


#c_values = np.arange(0.001,1,0.01)
#scores_c_logistic = {}
''''for c in c_values:
  model = LogisticRegression(solver='liblinear',C=c,random_state=1)
  model.fit(x_train,y_train)
  y_pred_val = model.predict_proba(x_val)[:,1]
  auc_train = roc_auc_score(y_train,y_pred_train)
  auc_val = roc_auc_score(y_val,y_pred_val)
  scores_c_logistic[c] = (auc_train,auc_val)
C_values = list(scores_c_logistic.keys())
scores = list(scores_c_logistic.values())

plt.figure(figsize=(10,5))
plt.plot(C_values, scores, marker='o')
plt.xlabel("C values (Inverse Regularization Strength)")
plt.ylabel("Accuracy")
plt.title("Logistic Regression Performance vs C")
plt.grid(True)
plt.show()'''
'''for i in range(1,11):
  scores_tree = {}
  tree = DecisionTreeClassifier(max_depth=i,random_state=1)
  tree.fit(x_train,y_train)
  y_pred_train = tree.predict_proba(x_train)[:,1]
  auc = roc_auc_score(y_train,y_pred_train)
  scores_tree['train'] = auc
  y_pred_val = tree.predict_proba(x_val)[:,1]
  auc = roc_auc_score(y_val,y_pred_val)
  scores_tree['val'] = auc
  y_pred_test = tree.predict_proba(x_test)[:,1]
  auc = roc_auc_score(y_test,y_pred_test)
  scores_tree['test'] = auc
  depth_score.append(scores_tree)'''


''''for i in range(1,11):
  estimator = {}
  random = RandomForestClassifier(n_estimators=10,random_state=1)
  random.fit(x_train,y_train)
  y_pred_train = random.predict_proba(x_train)[:,1]
  auc = roc_auc_score(y_train,y_pred_train)
  estimator['train'] = auc
  y_pred_val = random.predict_proba(x_val)[:,1]
  auc = roc_auc_score(y_val,y_pred_val)
  estimator['val'] = auc
  y_pred_test = tree.predict_proba(x_test)[:,1]
  auc = roc_auc_score(y_test,y_pred_test)
  estimator['test'] = auc
  estimator_socre.append(estimator)'''


#print(scores_logistic,"\n",tree_score,"\n",forest_score)
model = LogisticRegression(solver='liblinear',C=1.0,random_state=1)
model.fit(x_train,y_train)




```
